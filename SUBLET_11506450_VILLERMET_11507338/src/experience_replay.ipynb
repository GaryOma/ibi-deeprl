```python
def plot_reward(rewards, title="", file_name="error", mean_size=100, type_graph="dot"):
    list_e = []
    x_s = []
    i = 0
    if mean_size < 2:
        list_e = rewards
        x_s = range(len(rewards))
    else:
        while i < len(rewards):
            max_mean = i + mean_size
            if max_mean >= len(rewards):
                max_mean = len(rewards) - 1
            list_e.append(sum(rewards[i:max_mean])/(max_mean - i))
            x_s.append((max_mean + i)/2)
            i += mean_size
    # for error in errors:
    #     list_e.append(error.item())
    plt.figure(figsize=(30, 10))
    if type_graph == "dot":
        plt.plot(x_s, list_e, 'ok')
    else:
        # plt.bar(x_s, list_e)
        plt.plot(list_e, 'k')
    plt.ylabel("Récompense")
    plt.xlabel("Itération")
    plt.title(title)
    if file_name:
        plt.savefig(f"{SAV}{env_name}/{type_graph}_{file_name}.png")
    plt.show()
```

```python
class Experience:
    def __init__(self, _state, _action, _reward, _next_state, _done):
        self.state = _state
        self.action = _action
        self.reward = _reward
        self.next_state = _next_state
        self.done = _done
```

```python
class CNNet(nn.Module):
    def __init__(self, in_size, out_size):
        super(CNNet, self).__init__()
        self.outchannel1 = 32
        self.kernel1 = 8
        self.stride1 = 4
        print(in_size[0])
        print(out_size)
        self.conv1 = nn.Conv2d(in_size[0], self.outchannel1, self.kernel1, stride=self.stride1)
        # self.pool = nn.MaxPool2d(2, 2)
        self.outchannel2 = 64
        self.kernel2 = 4
        self.stride2 = 2
        self.conv2 = nn.Conv2d(self.outchannel1, self.outchannel2, self.kernel2, stride=self.stride2)
        self.outchannel3 = 64
        self.kernel3 = 3
        self.stride3 = 1
        self.conv3 = nn.Conv2d(self.outchannel2, self.outchannel3, self.kernel3, stride=self.stride3)
        # self.fc1 = nn.Linear(16 * self.outchannel2, 120)
        # self.fc2 = nn.Linear(120, 84)
        self.out3 = self.outchannel3 * 7 * 7
        self.fc = nn.Linear(self.out3, 512)
        self.fc2 = nn.Linear(512, out_size)

    def forward(self, x):
        x = x.unsqueeze(0)
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        x = x.view(-1, self.out3)
        x = F.relu(self.fc(x))
        x = self.fc2(x)
        return x.squeeze(0)
```

```python
class Net(nn.Module):
    def __init__(self, in_size, out_size):
        super(Net, self).__init__()
        print(in_size, out_size)
        # self.in_l = nn.Linear(in_size, out_size)
        self.in_l = nn.Linear(in_size, 3)
        self.in_l2 = nn.Linear(3, out_size)

    def forward(self, x):
        x = F.sigmoid(self.in_l(x))
        x = F.sigmoid(self.in_l2(x))
        return x
```

```python
class Agent:
    def __init__(self, _net, _in_size, _out_size, _target=None):
        self.mem = deque(maxlen=10000)
        with torch.no_grad():
            self.net = _net
            self.target = _target
            self.target_nn = _target
        self.gamma = 0.99
        self.in_size = _in_size
        self.out_size = _out_size
        self.beg_epsilon = 1
        self.end_epsilon = 0.1
        self.end_epsilon_step = 2000
        self.epsilon = self.beg_epsilon

    def choose_action(self, _observation, _reward, _done):
        qvalues = self.qvalue(_observation)
        action_ind = self.epsilongreedy(_observation, epsilon=self.epsilon)
        if self.epsilon > self.end_epsilon:
            self.epsilon *= (self.beg_epsilon - self.end_epsilon/self.end_epsilon_step)
        # action_ind = self.boltzmann(_observation, tau=0.1)
        return action_ind

    def experience(self, _state, _action, _reward, _next_state, _done):
        self.mem.append(Experience(_state, _action, _reward, _next_state, _done))

    def minibatch(self, size=5):
        batch_s = min(size, len(self.mem))
        return random.sample(self.mem, batch_s)

    def qvalue(self, state):
        inp = torch.tensor(state).type(torch.FloatTensor)
        out = self.net(inp)
        return out

    def qvalue_target(self, state):
        inp = torch.tensor(state).type(torch.FloatTensor)
        if self.target is not None:
            out = self.target(inp)
        else:
            out = self.net(inp)
        return out

    def epsilongreedy(self, ob, epsilon=0.5):
        if np.random.random() < epsilon:
            return np.random.randint(self.out_size)
        else:
            qval = self.qvalue(ob)
            _, max_ind = qval.max(0)
            return max_ind.item()

    def boltzmann(self, ob, tau=1000):
        qvalues = self.qvalue(ob)
        terms = torch.pow(np.e, torch.div(qvalues, tau))
        sum_ = terms.sum()
        probs = torch.div(terms, sum_)
        probs = np.array(probs)
        action_value = np.random.choice(probs,p=probs)
        action = np.argmax(probs == action_value)
        return action

    def sync_target(self):
        self.target = copy.deepcopy(self.net)
```

```python
def train(agent, env, episode_count, tot_ep=0):
    reward = 0
    rewards = []
    batch_calc = 0
    steps = 0
    for i in range(episode_count):
        ob = env.reset()
        rewardSum = 0
        done = False
        while not done:
            with torch.no_grad():
                action = agent.choose_action(ob, reward, done)
            current_state = ob
            if render:
                if (i + tot_ep) % episode_render == 0:
                    env.render()
                elif ((i + tot_ep) - 1) % episode_render == 0:
                    env.close()
            ob, reward, done, _ = env.step(action)
            rewardSum += reward
            steps += 1

            # Experience replay
            agent.experience(current_state, action, reward, ob, done)
            if done:
                print(f"ep {tot_ep + i} : reward {rewardSum}, step {steps} epsilon {agent.epsilon}", end="\r")
                rewards.append(rewardSum)

        minibatch = agent.minibatch(size=batch_size)
        for batch in minibatch:
            batch_calc += 1
            optimizer.zero_grad()
            qvalues = agent.qvalue(batch.state)
            with torch.no_grad():
                yj = batch.reward
                if not batch.done:
                    yj += agent.gamma * agent.qvalue_target(batch.next_state).max().item()
                real_y = qvalues.clone()
                new_q = (qvalues[batch.action].item() - yj) ** 2
                real_y[batch.action] = new_q
#                 print(f"real {real_y}")
#                 print(f"qvalues {qvalues}")
            loss = criterion(qvalues, real_y)
            loss.backward()
            optimizer.step()
            if batch_calc % target_sync == 0:
                agent.sync_target()
    return agent, rewards
```
